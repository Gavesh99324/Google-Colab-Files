```python
file_path="/content/drive/MyDrive/datamining_pratical2.csv"
df=pd.read_csv(file_path)

# Standardize 'Ma' to 'Male' and 'Fe' to 'Female', and convert to lowercase for consistent mapping
df['Gender'] = df['Gender'].replace({'Ma': 'Male', 'Fe': 'Female'})
df['Gender'] = df['Gender'].str.lower()

# Map standardized lowercase gender strings to numerical values
gender_mapping = {'male': 0, 'female': 1}
df['Gender'] = df['Gender'].map(gender_mapping)

print("Gender column standardized and encoded.")
display(df.head())
df.info()
```
